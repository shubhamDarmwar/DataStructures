## 1: Two Number Sum

```python
Image("Questions/Two Number Sum.png")
```

### Solutions
There are three ways to solve it:

1. Brute force algorithm using two nested for loops <br>
2. Using dictionry to store tracked numbers.<br>
3. Sorting array, and then using left right poiters

```python
def twoNumberSum(array, targetSum):
    
    left = 0
    right = len(array) - 1
    sortedArray = sorted(array)
    while(left < right):
        if (sortedArray[left] + sortedArray[right] == targetSum):
            return [sortedArray[left], sortedArray[right]]
        elif (sortedArray[left] + sortedArray[right] > targetSum):
            right -= 1
        else:
            left  += 1
    
    return []
```

```python
twoNumberSum([3,5,-4,8,11,1,-1,6], 10)
```

## 2. Validate Subsequence

```python
Image("Questions/validateSubsequence.png")
```

### Solution:
1. Using two loo- time: O(n2) + Space: O(1)<br>
2. Reducing one loop using map: time: O(n) + space: O(n)<br>
3. Using single loop on array. Trick is no need to search from start for each elemment. time: O(n) space: O(1)

```python
def validateSubsequence(array, sequence):
    seqIdx = 0
    for arrIdx in range(len(array)):
        if (seqIdx == len(sequence)):
            break
        if (sequence[seqIdx] == array[arrIdx]):
            seqIdx += 1
            
    return seqIdx == len(sequence)
```

```python
array = [5,1,22,25,6,-1,8,10]
sequence = [1,6,-1,10]

validateSubsequence(array, sequence)
```

## 3. Sorted Squared array

```python
Image("Questions/sortedSquaredArray.png")
```

### Solution

Trick: if -ve values are present then possible largerst square is of smallest or largest number. As array is sorted we can find absolute largest value. 

```python
def sortedSquaredArray(array):
    sIdx = 0
    eIdx = len(array) - 1
    sqArray = [0 for _ in array]
    idx = len(array) - 1

    for idx in reversed(range(len(array))):
        if (abs(array[sIdx]) > abs(array[eIdx])):
            sqArray[idx] = array[sIdx] * array[sIdx]
            sIdx += 1
        else:
            sqArray[idx] = array[eIdx] * array[eIdx]
            eIdx -= 1
    return sqArray
```

```python
sortedSquaredArray([ -7, -5, -4, 3, 6, 8, 9])
```

## 4. Tournament Winner

```python
Image('Questions/TournamentWinner1.png')
```

```python
def tournamentWinner(competitions, results):
	dic = {}
	for i in range(len(competitions)):
		homeTeam = competitions[i][0]
		awayTeam = competitions[i][1]
		if results[i] == 1:
			if homeTeam not in dic.keys():
				dic[homeTeam] = 1
			else:
				dic[homeTeam] = dic[homeTeam] + 1
		else:
			if awayTeam not in dic.keys():
				dic[awayTeam] = 1
			else:
				dic[awayTeam] = dic[awayTeam] + 1
	winner = None
	for key in dic.keys():
		if winner == None:
			winner = key
		elif dic[winner] < dic[key]:
			winner = key
	
	return winner
```

```python
competitions = [["HTML","CSS"], ["CSS", "PYTHON"], ["PYTHON", "HTML"]]
results = [0,0,1]
tournamentWinner(competitions, results)
```

## 5. Non-Constuctible Minimum change

```python
Image("Questions/5_NonConstructibleChange.png")
```

### Solution:
Given:<br>
1. You can not use single coin multiple times to create change<br>
2. Non -ve

Trick:<br>
1. Sort array<br>
2. As we need minimum non constructible amount start summing smaller coins<br>
3. At each iteration coin should not be greater than (sum + 1). If it false we can not create change (sum + 1)<br>

```python
def nonConstructibleChange(coins):
	coins.sort()
	sum_ = 1
	for coin in coins:
		if coin > sum_:
			return sum_
		
		sum_ += coin
	return sum_
```

```python
coins = [5,7,1,1,2,3,22]
nonConstructibleChange(coins)
```

## 6. Closest value in BST

```python
def findClosestValueInBst(tree, target):
    closest = tree.value 
    currentNode = tree
    while currentNode is not None:
        if abs(target - closest) > abs(target - currentNode.value):
            closest = currentNode.value;
        if target < currentNode.value:
            currentNode = currentNode.left
        elif target > currentNode.value:
            currentNode = currentNode.right
        else:
            break
    return closest

# This is the class of the input tree. Do not edit.
class BST:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
```

```python
def branchSums(root):
    sums = []
    calculateSums(root, 0, sums)
    return sums

def calculateSums(node, runningSum, sums):
	if node is None:
		return
	newRunningSum = runningSum + node.value
	if node.left is None and node.right is None:
		sums.append(newRunningSum)
		return
	calculateSums(node.left, newRunningSum, sums)
	calculateSums(node.right, newRunningSum, sums)
```

```python
def nodeDepths(root):
    sumDepth = 0
    stack = [{"node": root, "depth": 0}]
    while len(stack) > 0:
        info = stack.pop()
        node, depth = info["node"], info["depth"]
    
        if node is None:
            continue
        sumDepth += depth
        stack.append({"node": node.left, "depth": depth + 1})
        stack.append({"node": node.right, "depth": depth + 1})
    return sumDepth
```

## 9. Depth first search

```python
Image("Questions/9_DeptFirstSearch.png")
```

### Solution 1:

```python
class Node:
    def __init__(self, name):
        self.children = []
        self.name = name

    def addChild(self, name):
        self.children.append(Node(name))
        return self
```

```python
def depthFirstSearch(root, array):
        stack = [];
        cptr = root;
        while cptr is not None:
            array.append(cptr.name)
            stack.append({"node" : cptr, "ctt" : 1})
            if len(cptr.children) > 0:
                cptr = cptr.children[0]
            else:
                cptr = None;
                
        while len(stack) > 0:
            element = stack.pop()
            node, ctt = element["node"], element["ctt"]

            if len(node.children) > ctt:
                stack.append({"node" : node, "ctt" : ctt + 1})
                node = node.children[ctt]
                array.append(node.name)
                stack.append({"node" : node, "ctt" : 0})
                
        return array
```

## 10. Miimum waiting time (Greedy algorithm)
This is called greedy algorithm because we are making greedy choise for which query to execute first.

```python
Image("Questions/10_MinWiatingTime.png")
```

### Solution 1:
Sort and then cumulative sum.

Time:  O(nlog(n)) Space : O(1)

```python
def minimumWaitingTime(queries):
    queries.sort();
    twt = 0;
    wt = 0;
    for i in queries:
        twt += wt;
        wt += i;
    return twt;
```

### Solution 2:

1. Sort<br>
2. loop through<br>
3. sum waiting time for all remaining queries due to current query<br>

Time:  O(nlog(n)) Space : O(1)

```python
def minimumWaitingTime(queries):
    queries.sort();
    twt = 0
    for idx, t in enumerate(queries):
        queriesLeft = len(queries) - (idx + 1)
        twt += t * queriesLeft
    return twt
```

```python
def classPhotos(redShirtHeights, blueShirtHeights):
    redShirtHeights.sort(reverse=True);
    blueShirtHeights.sort(reverse=True);
    
    blueAtBack = False;

    if redShirtHeights[0] < blueShirtHeights[0]:
        blueAtBack = True
    
    for i in range(len(redShirtHeights)):
        if blueAtBack:
            if redShirtHeights[i] >= blueShirtHeights[i]:
                return False
        else:
            if redShirtHeights[i] <= blueShirtHeights[i]:
                return False
    return True
```

### Solution:
Time: O(n) Space: O(1)

```python
def tandemBicycle(redShirtSpeeds, blueShirtSpeeds, fastest):
    blueShirtSpeeds.sort()
    if fastest:
        redShirtSpeeds.sort(reverse=True)
    else:
        redShirtSpeeds.sort();

    totalSpeed = 0;
    for i in range(len(redShirtSpeeds)):
        totalSpeed += max(redShirtSpeeds[i], blueShirtSpeeds[i])
        
    return totalSpeed;
```

## 13 Remove Duplicates from Linked list

```python
Image('Questions/13_RemoveDuplicatesLinkedList.png')
```

### Solution:
Time: O(n) Space: O(1)

```python
# This is an input class. Do not edit.
class LinkedList:
    def __init__(self, value):
        self.value = value
        self.next = None


def removeDuplicatesFromLinkedList(linkedList):
    dic = {};
    cptr = linkedList;
    oldCptr = None
    while (cptr is not None):
        if cptr.value not in dic:
            dic[cptr.value] = 1;
            oldCptr = cptr
        else:
            oldCptr.next  = cptr.next;

        cptr = cptr.next
        
    return linkedList;
```

## 14 Nth  Fibonacci number

```python
Image('Questions/14_Nth_fibonacci.png')
```

### Solution 1:

#### time: O(2^n) 
At each time two new functions getting called

#### space: O(n)
At any give time maximum call stack size is less than n


```python
def getNthFib(n):
    if n == 2:
        return 1;
    elif n ==  1:
        return 0
    return getNthFib(n-1)  + getNthFib(n-2)
```

### Solution 2:

#### time: O(n)

#### space: O(n)

```python
def getNthFib(n, memoize={1:0, 2:1}):
    if n in memoize:
        return memoize[n]
    else:
        memoize[n] = getNthFib(n-1, memoize) + getNthFib(n-2, memoize)
        return memoize[n]
```

### Solution 3:
time: O(n) space: O(1)

```python
def getNthFib(n):
    if n  == 1:
        return 0
    elif n == 2:
        return 1
    n1 = 0
    n2 = 1
    counter = 3
    while counter <= n:
        nextFib = n1 + n2;
        n1 = n2
        n2 = nextFib
        counter += 1
    return n2;
```
