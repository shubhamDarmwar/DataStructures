## Linked List Implementation

```python
class Node:
    def __init__(self):
        self.data = None
        self.next = None
        self.randomPointer = None
        
    def __del__(self):
        pass
#         print('Deleted {}'.format(self.data))

        
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def append(self, data):
        newNode = Node()
        newNode.data = data
        newNode.next = None
        
        if self.head == None:
            self.head = newNode
            self.tail = newNode
            
        else:
            self.tail.next = newNode
            self.tail = self.tail.next
        
    
    def insertAtStart(self, data):
        newNode = Node()
        newNode.data = data
        newNode.next = None
        
        if self.head == None:
            self.head = newNode
            self.tail = newNode
            
        else:
            newNode.next = self.head
            self.head = newNode
            
    def addLoopAt(self, position):
        index = 1
        cptr = self.head
        while cptr.next != None and index < position:
            cptr = cptr.next
            index += 1
        self.tail.next = cptr
        self.tail = cptr
        
    def addRandomLink(self, source, destination):
        index = 1
        cptr = self.head
        while index != source:
            index += 1
            cptr = cptr.next
        sourceNode = cptr
        
        cptr = self.head
        index = 1 
        while index != destination:
            index += 1
            cptr = cptr.next
        destinationNode = cptr
        sourceNode.randomPointer = destinationNode
        
        print(f"{sourceNode.data} {sourceNode.randomPointer.data}")
        
    def deleteHead(self):
        self.head = None
        
    def printNodes(self):
        cptr = self.head
        while cptr != None:
            print(cptr.data, end="")
            print(" -> ", end="") if cptr.randomPointer == None else print(f"({cptr.randomPointer.data})", end=" -> ")
            cptr = cptr.next
        print(" ")
        
```

### Stack using LinkedList

```python
class LStack:
    def __init__(self):
        self.array = LinkedList()
        
    def push(self, data):
        self.array.insertAtStart(data)
    
    def pop(self):
        if self.array.head != None:
            data = self.array.head.data
            self.array.head = self.array.head.next
            return  data
        
```

### Problem 3:
#### Reverse Linked List

#### Solutions:
1. Using stack
2. Recursive
3. Reverse list inplace

```python
def reverse(head):
    if head == None or head.next == None:
        return head
    
    firstptr = head
    secondptr = head.next
    thirdptr = head.next.next
    
    firstptr.next = None
    
    while True:
        if thirdptr == None:
            secondptr.next = firstptr
            return secondptr
        secondptr.next = firstptr
        firstptr = secondptr
        secondptr = thirdptr
        thirdptr = thirdptr.next
    
```

### Problem 6:
#### Merge two sorted linked list (Recursive)

```python
def mergeSortedLists(head1, head2):
    resultHead = None
    if head1 == None:
        return head2
    if head2 == None:
        return head1
    if head1.data <= head2.data:
        resultHead = head1
        resultHead.next = mergeSortedLists(head1.next, head2)
    else:
        resultHead = head2
        resultHead.next = mergeSortedLists(head1, head2.next)
    
    return resultHead
    
```

### Problem 7:
#### Reverse linked list in pairs
L1: 1, 2, 3, 4, 5<br>
O/P: 2, 1, 4, 3, 5

```python
def reverseLinkedListInPairs(head):
    if head == None or head.next == None:
        return head
    ptr1 = head
    ptr2 = head.next
    ptr3 = head.next.next
    tail = None
    
    newHead = ptr2
    
    while not (ptr2 == None and ptr3 == None):
            
        ptr1.next = ptr3
        ptr2.next = ptr1
        if tail != None:
            tail.next = ptr2
        tail = ptr1
        if ptr3 == None:
            return newHead
        
        ptr1 = ptr3
        ptr2 = ptr1.next
        if ptr2 != None:
            ptr3 = ptr2.next
        else:
            ptr3 = None

    return newHead
            
```

```python
def reverseInPairsRecursive(head):
    if head == None or head.next == None:
        return head
    
    second = head.next
    temp = second.next
    
    second.next = head
    head.next = reverseInPairsRecursive(temp)
    
    return second
```

### Problem 10:
#### Clone linked list. Each node have next and random pointer.
#### Solution:
In first iteration create new nodes and store allin hash table agains key as a original node.<br>
Second Iteration: Fill new node's next and random pointers<br>
Time: O(n)
Space: O(n)<br>

```python
def cloneLL(head):
    if head == None:
        return
    hashtable = {}
    cptr = head
    resultHead = None
    
    while cptr != None:
        cloneNode = Node()
        cloneNode.data = cptr.data
        hashtable[cptr] = cloneNode
        cptr = cptr.next
    cptr = head
    while cptr != None:
        
        cloneNode = hashtable[cptr]
        if cptr.next != None:
            cloneNode.next = hashtable[cptr.next]
        if cptr.randomPointer != None:
            cloneNode.randomPointer = hashtable[cptr.randomPointer]
        
        if resultHead == None:
            resultHead = cloneNode
        cptr = cptr.next
    return resultHead
        

```

### Problem 19
Given a sorted linked list, delete all duplicates such that each element appear only once.<br>

For example,<br>
Given 1->1->2, return 1->2.<br>
Given 1->1->2->3->3, return 1->2->3.

VMWare, Microsoft

```python
def deleteDuplicates(A):
    cptr = A
    head = A
    oldCptr = None
    while cptr != None:

        if oldCptr != None and cptr.data == oldCptr.data:

            oldCptr.next = cptr.next
            cptr.next = None
            cptr = oldCptr.next
        else:
            oldCptr = cptr
            cptr = cptr.next

    return head
```

### Problem 21 : Merge Two Sorted Lists

Merge two sorted linked lists and return it as a new list. <br>
The new list should be made by splicing together the nodes of the first two lists, and should also be sorted.

Amazon, Microsoft, Yahoo

```python
def mergeTwoLists(A, B):
    head = None
    a = A
    b = B
    H = None
    while True:
        if head == None:
            if a.data >= b.data:
                head = b
                temp = b.next
                b.next = None
                b = temp
            else:
                head = a
                temp = a.next
                a.next = None
                a = temp
            H = head
        else:
            if a.data >= b.data:
                head.next = b
                head = head.next
                temp = b.next
                b.next = None
                b = temp
            else:
                head.next = a
                head = head.next
                temp = a.next
                a.next = None
                a = temp

        if a == None:
            head.next = b
            return H
        elif b == None:
            head.next = a
            return H
```

### Problem 26:  List Cycle
Given a linked list, return the node where the cycle begins. If there is no cycle, return null.

Try solving it using constant additional space.

Example :

Input :

                  ______
                 |     |
                 \/    |
        1 -> 2 -> 3 -> 4

Return the node corresponding to node 3. 

Amazon, Microsoft, NetApp

```python
# @param A : head node of linked list
# @return the first node in the cycle in the linked list
def detectCycle( A):
    slow = A
    fast = A

    while fast != None:
        slow = slow.next
        if fast.next != None:
            fast = fast.next.next
        else:
            fast == None
            return None
        if slow == fast:
            break

    if fast == None:
        return None
    slow = A
    while slow != fast: ## Assumption is both the pointers at the same distance
        slow = slow.next
        fast = fast.next

    return slow
```
